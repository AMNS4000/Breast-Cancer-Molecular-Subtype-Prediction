```python
# Convert each entry of SeriesDataRole column to lowercase
df_clean['SeriesDataRole'] = df_clean['SeriesDataRole'].str.lower()

# Keep only rows with SeriesDataRole as 'post'
df_clean = df_clean[df_clean['SeriesDataRole'] == 'post']

# reindex the dataframe
df_clean = df_clean.reset_index(drop=True)
print(df_clean.shape)
df_clean.head()
```

```python
# drop rows where PatientID is duplicated
df_clean = df_clean.drop_duplicates(subset='PatientID', keep='first')
df_clean = df_clean.reset_index(drop=True)
print(df_clean.shape)
df_clean.head()
```

```python
# check if there are any missing values in the dataframe using a for loop
for column in df_clean.columns:
    if df_clean[column].isnull().sum() > 0:
        print(f"{column}: {df_clean[column].isnull().sum()}")
```

```python
# convert each entry of PatientAge column to int by removing 'Y' from the end of each entry and ignore the missing values
for i in range(df_clean.shape[0]):
    if pd.notnull(df_clean['PatientAge'][i]):
        df_clean['PatientAge'][i] = int(df_clean['PatientAge'][i][:-1])

# fill the missing values in patientage column with the mean of the column
df_clean['PatientAge'] = df_clean['PatientAge'].fillna(df_clean['PatientAge'].mean())
df_clean['PatientAge'].dtype
```

```python
# print each datatype of each column
object_columns = []
for column in df_clean.columns:
    if df_clean[column].dtype == 'object':
        print(f"{column}: {df_clean[column].dtype}")
        if column not in ['PatientID', 'SeriesDataRole']:
            object_columns.append(column)
```

```python
# Perform label encoding on the object columns
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for column in object_columns:
    if column in ['PatientID', 'SeriesDataRole']:
        continue
    if df_clean[column].dtype == 'object':
        df_clean[column] = label_encoder.fit_transform(df_clean[column])

print(df_clean.shape)
df_clean.head()
```

```python
# convert each datatype to float and int
for column in df_clean.columns:
    if column in ['PatientID', 'SeriesDataRole']:
        continue
    if df_clean[column].dtype == 'object':
        df_clean[column] = df_clean[column].astype(float)
df_clean
```

```python
# Export the csv file
df_clean.to_csv('../../Without Demographic Features/Data/MAX_features.csv', index=False)
```
